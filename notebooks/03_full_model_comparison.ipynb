# Full Model Comparison

Load the saved comparison table and visualize equity curves plus Sharpe ratios across all configured models.

```python
from pathlib import Path
import pandas as pd
from matplotlib import pyplot as plt

results_dir = Path('../results')
comparison = pd.read_csv(results_dir / 'model_comparison.csv').sort_values('sharpe', ascending=False)
display_names = {
    'xgboost_fallback_histgb': 'XGBoost Fallback (HistGB)',
    'lightgbm_fallback_histgb': 'LightGBM Fallback (HistGB)',
    'buy_and_hold_spy': 'Buy & Hold SPY',
    'lstm': 'LSTM',
    'temporal_fusion_transformer': 'Temporal Fusion Transformer',
    'patchtst': 'PatchTST',
    'itransformer': 'iTransformer',
    'mamba': 'Mamba',
}
comparison['display_name'] = comparison['model'].map(display_names).fillna(comparison['model'])
comparison
```

```python
figure, axis = plt.subplots(figsize=(11, 4))
colors = ['#1f77b4' if model != 'buy_and_hold_spy' else '#2ca02c' for model in comparison['model']]
axis.bar(comparison['display_name'], comparison['sharpe'], color=colors)
axis.set_title('Sharpe Ratio by Model')
axis.set_ylabel('Sharpe')
axis.tick_params(axis='x', rotation=45)
axis.grid(axis='y', alpha=0.3)
figure.tight_layout()
```

```python
figure, axis = plt.subplots(figsize=(12, 5))
for model_name in comparison['model']:
    curve_path = results_dir / 'equity_curves' / f'{model_name}_aggregated.csv'
    if curve_path.exists():
        curve = pd.read_csv(curve_path, parse_dates=['date'])
        label = display_names.get(model_name, model_name)
        linewidth = 2.2 if model_name == 'buy_and_hold_spy' else 1.4
        axis.plot(curve['date'], curve['equity_curve'], label=label, linewidth=linewidth)

axis.set_title('Aggregated Equity Curves')
axis.set_xlabel('Date')
axis.set_ylabel('Equity')
axis.grid(alpha=0.3)
axis.legend(ncol=2)
figure.tight_layout()
```
